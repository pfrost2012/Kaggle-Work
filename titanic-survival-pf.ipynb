```python
train_data = pd.read_csv('/kaggle/input/titanic/train.csv')
test_data = pd.read_csv('/kaggle/input/titanic/test.csv')
train_data
```

```python
train_data['Sex']= train_data['Sex'].str.lower().map({'female':0, 'male':1})
test_data['Sex'] = test_data['Sex'].str.lower().map({'female':0, 'male':1})
train_data
```

```python
train_data = train_data.select_dtypes(include=['int64','float64'])
test_data = test_data.select_dtypes(include=['int64','float64'])
```

```python
train_data = train_data.dropna()
```

```python
train_data
```

```python
test_data.isna().any()
```

```python
from sklearn.model_selection import train_test_split
```

```python
features = ['PassengerId','Pclass','Sex','Parch','SibSp']
y = train_data['Survived']
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])
```

```python
from sklearn.ensemble import RandomForestClassifier
lr = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
lr.fit(X,y)
predictions = lr.predict(X_test)
```

```python
predictions
```

```python
df_predictions = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
```
